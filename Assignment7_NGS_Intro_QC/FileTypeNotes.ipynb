### BAM files

- compressed version of SAM

- bam file is a binary so wer cant read and contains unmapped reads, multiple alignments on the same read and suppulentary reads  

### SAM Files

- sam is tab delimted for storing read alignments to a reference genome 

- One record (a single DNA fragment alignment) per line describing alignment between fragmetns and reference 

- 11 fixed colums + optional key:type:value tuples

- 12 colums listed in SAM file:

    > Query Name 
    
    > Bitwise FLAG
    
    > reference sequence name 
    
    > position 1-bated left most position of clipped alignment 
    
    >mapping quality (needs to be greater than 20 for a good score)
    
    >extended CIGAR string (if the nucleotide is a match/instertion/deletion)
    
    >Mate reference NaMe
    
    >1-based leftmost mate position 
    
    >infered insert size
    
    >query sequence on the dame strand as the reference 
    
    >query quality 
    
    >optional fields 
    

- Flags are the number of bits (12 bit long strings ) representing the flags present of ) 0= false , 1=true
